```python
import numpy as np
import matplotlib.pyplot as plt

# 두 직선의 계수를 정의합니다
# 첫 번째 직선: 2x + 3y = 5
# 두 번째 직선: -x + y = 5
A = np.array([[2, 3], [-1, 1]])
B = np.array([5, 5])

# 교점 계산
intersection = np.linalg.solve(A, B) #선형대수(linear algebra)의 해를 구한다(solve) 
x_intersection, y_intersection = intersection

# 직선을 그리기 위해 x 범위 설정
x_values = np.linspace(-10, 10, 100)

# 두 직선의 y 값 계산
y1_values = (5 - 2 * x_values) / 3
y2_values = 5 + x_values

# 시각화
plt.figure(figsize=(8, 6))
plt.plot(x_values, y1_values, label="2x + 3y = 5")
plt.plot(x_values, y2_values, label="-x + y = 5")
plt.scatter(x_intersection, y_intersection, color='red', zorder=5, label=f'Intersection ({x_intersection:.2f}, {y_intersection:.2f})')

plt.xlabel("x")
plt.ylabel("y")
plt.title("Intersection of Two Lines")
plt.axhline(0, color='grey', lw=0.5)
plt.axvline(0, color='grey', lw=0.5)
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np

# 세 개의 3차원 벡터 정의
A_three_dim = np.array([1, 0, 2])
B_three_dim = np.array([0, 1, 3])
C_three_dim = np.array([1, 1, 1])

# 목표 벡터
target_vector = np.array([9, 11, 2])

# 선형 결합 계수 계산
coeff_matrix = np.column_stack([A_three_dim, B_three_dim, C_three_dim])
intersection_b = np.linalg.solve(coeff_matrix, target_vector)

# zip을 이용하여 선형 결합 계산
linear_combination = sum(i * j for i, j in zip([A_three_dim, B_three_dim, C_three_dim], intersection_b))

# 출력
print("계산된 선형 결합 계수 (intersection_b):", intersection_b)
print("선형 결합 결과:", linear_combination)
print("원래 목표 벡터:", target_vector)
print("일치 여부:", np.allclose(linear_combination, target_vector))
```
